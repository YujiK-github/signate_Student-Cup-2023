```python
import pandas as pd
import folium
df = pd.read_csv("G:/マイドライブ/signate_StudentCup2023/data/state_coordinate.csv")
df.head()
```

```python
# 地図の中心を計算
center_lat = df['state_latitude'].mean()
center_lon = df['state_longitude'].mean()

# 地図オブジェクトを作成
m = folium.Map(location=[center_lat, center_lon], zoom_start=4)

# データを地図上に表示
for index, row in df.iterrows():
    folium.Marker([row['state_latitude'], row['state_longitude']], popup=row['state']).add_to(m)
m.save("state_map.html")
```
