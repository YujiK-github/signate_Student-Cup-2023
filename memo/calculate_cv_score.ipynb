```python
# ===================================================================
#  Library
# ===================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
```

```python
# ===================================================================
#  data
# ===================================================================
data = pd.DataFrame(
    data=zip(
        [45.49498124624943, 
         44.364568837517524,
         44.706570976834534,
         44.33097999776283,
         44.33737773847836,
         ],
        [44.7311468, 
         43.7384163,
         44.0149739,
         43.7719024,
         43.7822822,
         ],
        ),
    columns=["CV", "LB"]
)
print(sp.stats.pearsonr(data["LB"], data["CV"])[0])
data
```

```python
# ===================================================================
#  plot
# ===================================================================
plt.scatter(x=data["LB"], y=data["CV"])

# Perform linear regression
coefficients = np.polyfit(data["LB"], data["CV"], 1)
poly_eq = np.poly1d(coefficients)

# Create x values for the line of best fit
x_fit = np.linspace(min(data["LB"]), max(data["LB"]), 100)

# Plot the line of best fit
plt.plot(x_fit, poly_eq(x_fit), color="dodgerblue", alpha=0.25)

# Add labels, title, legend, and grid
plt.xlabel("LB Values")
plt.ylabel("CV Values")
plt.title("Scatter Plot of LB vs. CV with Approximate curve")
plt.grid()

# Display the plot
plt.show()
```

```python
# ===================================================================
#  CVの計算
# ===================================================================
best_score = 43.0831209
poly_eq([best_score])[0]
```
